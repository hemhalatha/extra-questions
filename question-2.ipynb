```python
import random
import matplotlib.pyplot as plt

def linear_search(arr, target):
    comparisons = 0
    for i in range(len(arr)):
        comparisons += 1
        if arr[i] == target:
            return comparisons
    return comparisons

input_sizes = [100, 500, 1000, 5000, 10000]

best_case = []
average_case = []
worst_case = []

for n in input_sizes:
    arr = list(range(1, n + 1))

    best = linear_search(arr, arr[0])
    best_case.append(best)

    avg_comparisons = 0
    for _ in range(100):
        target = random.choice(arr)
        avg_comparisons += linear_search(arr, target)
    average_case.append(avg_comparisons // 100)

    worst = linear_search(arr, arr[-1])
    worst_case.append(worst)

plt.plot(input_sizes, best_case, label="Best Case (O(1))", marker='o')
plt.plot(input_sizes, average_case, label="Average Case (O(n/2))", marker='o')
plt.plot(input_sizes, worst_case, label="Worst Case (O(n))", marker='o')

plt.xlabel("Input Size (n)")
plt.ylabel("Number of Comparisons")
plt.title("Linear Search - Number of Comparisons")
plt.legend()
plt.grid(True)
plt.show()
```
